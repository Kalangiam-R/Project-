```python
import pandas as pd
df = pd.read_csv('/content/HR-Employee-Attrition.csv')
print(df.head())
```

```python
import matplotlib.pyplot as plt

experience_attrition = df.groupby('TotalWorkingYears')['Attrition'].value_counts().unstack()
experience_attrition.plot(kind='bar',figsize=(10,5))
plt.title("Attrition by Experience")
plt.xlabel("years of Experience")
plt.ylabel("Employee")
plt.show()
```
